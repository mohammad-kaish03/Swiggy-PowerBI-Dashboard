# data preproccessing

## missing value

```python
df.isnull().sum()
```

```python
df.isnull().sum()/len(df)*100
```

```python
df["Item Weight"].fillna(df["Item Weight"].median(),inplace=True)
df["Item Visibility"].fillna(df["Item Visibility"].median(),inplace=True)
```

```python
df.isnull().sum()
```

### observation: 
- we got 1463 missing values as null in Item Weight (17%)
- we got 23 missing values as null in Item Visibility (0.26%)
- we replaced null values with median of the column

## duplicate data

```python
df.duplicated().sum()
```

### observation:
- there was no duplicate records

## wrong filling in fat contant

```python
df["Fat Content"].unique()
```

```python
df["Fat Content"] = df["Fat Content"].replace('low fat', 'Low Fat')
df["Fat Content"] = df["Fat Content"].replace('reg', 'Regular')
df["Fat Content"] = df["Fat Content"].replace('LF', 'Low Fat')
```

### observation:
#### Fat Content
- it has 5 unique values
- 3 of them was miss-spelled
- we replaced them with 2 unique values
